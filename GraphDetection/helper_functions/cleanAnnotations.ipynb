```python
# Go through and get the image sizes of each image, and store it under its name.\
import os
import sys

# This is where we need to store the final labels.
training_labels = "Roboflow/train/label.txt"
val_labels = "Roboflow/valid/label.txt"

train_images_dir = "Roboflow/train/images"
train_labels_dir = "Roboflow/train/labels"

val_images_dir = "Roboflow/valid/images"
val_labels_dir = "Roboflow/valid/labels"

for filename in os.listdir(train_images_dir):
    file = filename.split(".")[0]
    extension = filename.split(".")[-1]

    if extension != "jpg":
        continue

    # Use the full filename except with .txt to get the annotations.
    annotations_file = train_labels_dir + '/' + '.'.join(filename.split('.')[0:-1]) + '.txt'

    # Open the annotations file, read it.
    # Each row represents a new graph, unless it is the comment with the file name.
    ann_lines = []
    with open(annotations_file, 'r') as af:
        ann_lines = af.readlines()
        
    
    if len(ann_lines) > 0:
        last_element = ann_lines[-1]
        ann_lines[-1] = last_element + '\n'
    
    
    # Each row is in the form 0 x_center y_center width height. 
    
    # Open the labels text file and write the rows.
    # Write one line for the filename.
    # ann_lines = [f'# {filename}\n'] + ann_lines

    with open(training_labels, 'a') as tl:
        tl.write(f'# {filename}\n')
        tl.writelines(ann_lines)



for filename in os.listdir(val_images_dir):
    file = filename.split(".")[0]
    extension = filename.split(".")[-1]

    if extension != "jpg":
        continue

    # Use the full filename except with .txt to get the annotations.
    annotations_file = val_labels_dir + '/' + '.'.join(filename.split('.')[0:-1]) + '.txt'

    # Open the annotations file, read it.
    # Each row represents a new graph, unless it is the comment with the file name.
    ann_lines = []
    with open(annotations_file, 'r') as af:
        ann_lines = af.readlines()
    
    if len(ann_lines) > 0:
        last_element = ann_lines[-1]
        ann_lines[-1] = last_element + '\n'
    
    # Each row is in the form 0 x_center y_center width height. 
    
    # Open the labels text file and write the rows.
    # Write one line for the filename.
    # ann_lines = [f'# {filename}\n'] + ann_lines
    with open(val_labels, 'a') as vl:
        vl.write(f'# {filename}\n')
        vl.writelines(ann_lines)
    






```
